```python
from keras.utils import pad_sequences
from pickle import dump,load
from keras.preprocessing.text import Tokenizer
from keras.utils import to_categorical
import numpy as np
```

```python
def readNames(fileName):
    return open(fileName,'r',encoding='utf8').readlines()
```

```python
dataNames = readNames('male.txt')
```

```python
def generateSequence():
    results = []
    for name in dataNames:
        for index in range(2,len(name)):
            results.append(name[:index+1])
    return np.array(results)
```

```python
dataNames = generateSequence()
```

```python
vocabSize = len(tokenizer.word_index) + 1
seqLength = xFeatures.shape[1]
seqLength,vocabSize
```

```python
yLabelsBinary = to_categorical(yLabels,num_classes=vocabSize)
```

```python
from keras.models import Sequential
from keras.layers import LSTM,Bidirectional,Dense,Embedding
```

```python
def createModel(vocabSize,seqLength):
    model = Sequential([
        Embedding(vocabSize,seqLength,input_length=seqLength),
        Bidirectional(LSTM(250,return_sequences=True)),
        Bidirectional(LSTM(200,return_sequences=True)),
        LSTM(150),
        Dense(100,activation='relu'),
        Dense(50,activation='relu'),
        Dense(vocabSize,activation='softmax')
    ])
    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    model.summary()
    return model
```

```python
model = createModel(vocabSize,seqLength)
```

```python
history = model.fit(
                    xFeatures,
                    yLabelsBinary,
                    batch_size=32,
                    epochs=50,
                    verbose=True
                   )
```

```python
model.save('textGeneration.h5')
dump(tokenizer,open('my_simpleTokenizer','wb'))
dump(history,open('model_history','wb'))
```

```python
def convertToText(model,seedText,tokenizer,N):
    for i in range(0,N):
        # converting the seedText to tokens
        newText = seedText
        wordIndex = tokenizer.word_index
        processed = np.array([wordIndex[word] for word in newText if word in wordIndex])
        processed = pad_sequences([processed],maxlen=15,padding='pre')
        # getting predictions
        prediction = model.predict(processed,verbose=False)[0]
        index = np.argmax(prediction)
#         wordIndex = yLabels[index]
        actualWord = tokenizer.index_word[index]
        seedText += actualWord
    return seedText
```
